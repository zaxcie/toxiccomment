```python
import pandas as pd

from keras.preprocessing import text
```

```python
max_features = 50000
maxlen = 200

train = pd.read_csv("data/raw/train.csv")
test = pd.read_csv("data/raw/test.csv")
train = train.sample(frac=1)

list_sentences_train = train["comment_text"].fillna("CVxTz").values
list_classes = ["toxic", "severe_toxic", "obscene", "threat",
                "insult", "identity_hate"]
y = train[list_classes].values
list_sentences_test = test["comment_text"].fillna("CVxTz").values


tokenizer = text.Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(list(list_sentences_train))
```

```python
tokenizer.word_counts

import operator
sorted_x = sorted(tokenizer.word_counts.items(), key=operator.itemgetter(1), reverse=True)
```

```python
popular_word = dict()

for i in list_classes:
    list_sentences = train[(train[i] == 1)]["comment_text"].fillna("CVxTz").values
    token = text.Tokenizer(num_words=max_features)
    token.fit_on_texts(list(list_sentences))
    
    sorted_x_token = sorted(token.word_counts.items(),
                      key=operator.itemgetter(1),
                      reverse=True)
    
    popular_word[i] = sorted_x_token
```
