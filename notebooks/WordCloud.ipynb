```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

### Import and format data

```python
train = pd.read_csv("data/raw/train.csv")
classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
```

### Whole dataset Word Cloud

```python
corpus = train["comment_text"].str.cat(sep=" ")

wordcloud = WordCloud(max_font_size=35).generate(corpus)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### Word cloud for a specific classe

```python
for i in classes:
    corpus = train[(train[i] == 1)]["comment_text"].str.cat(sep=" ")
    
    wordcloud = WordCloud(max_font_size=35).generate(corpus)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(i)
    plt.show()
```
